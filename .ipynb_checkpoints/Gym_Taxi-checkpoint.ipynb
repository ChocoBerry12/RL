# Taxi
---
> 택시 도메인. RGBY 4 개 블록 중 하나에 승객이 있고 다른 하나의 블록이 목적지이다. 택시는 랜덤 위치에서 시작한다. 원래 계층적 강화학습이나 관계형 강화학습 연구논문에서 처음 제안된 도메인인데 간단한 Q 러닝으로 연습용으로 풀기 좋은 것 같다.

* state : 25(택시 위치 = map 크기) x 5(승객 위치 - R, G, B, Y, taxi) x 4(목적지 - R, G, B, Y)
* actoin : 하-0 상-1 우-2 좌-3 태우기-4 내리기-5
* reward : 매 step 마다 -1, 제대로 운송시 +20, 잘못 내리기 또는 태우기 할 때 -10
* rendering : 파란색-승객, 마젠타-목적지, 노란색-빈 택시, 초록색-승객이 탄 택시
* etc : 200 step 초과하면 종료
---

```python
import gym
import numpy as np
import random
```

```python
'''
환경 생성
'''
env = gym.make('Taxi-v2')
```

```python
print(env.action_space.n)
print(env.observation_space.n)
```

```python
q_table = np.zeros([env.action_space.n, env.observation_space.n])
gamma = .95
epsilon = 1
alpha = .1
episode_total = 3000
episode = 0
```

```python
state = env.reset()
step = 0

while(True):
    #env.render()
    step += 1
    action = env.action_space.sample()
    state_next, reward, done, _ = env.step(action)
    if(done):
        env.reset()
        print(reward, step)
        if(reward > 0):
            break
        step = 0
env.close()
```

```python
state = env.reset()
step = 0

while(episode < episode_total):
    step += 1
    if(random.random() < epsilon):
        action = env.action_space.sample()
    else:
        action = np.argmax(q_table[ : , state])

    state_next, reward, done, _ = env.step(action)
    q_table[action, state] += alpha * (reward + np.max(gamma * q_table[ : , state_next]) - q_table[action, state])
    state = state_next

    if(done):
        if(step == 200):
            # 에피소드 증가 ㄴㄴ, 그냥 버리기
            step = 0
            continue
            
        if(reward > 0):
            #print(reward)
            if(epsilon < .1):
                epsilon = .1
            else:
                epsilon = 1 / (1 + episode / 10)
        episode += 1
        env.reset()
        step = 0
env.close()
```

```python
print(q_table)
s = env.reset()

while(True):
    a = np.argmax(q_table[ : , s])
    s,r,d,_ = env.step(a)
    env.render()
    if(d):
        env.render()
        break
```
